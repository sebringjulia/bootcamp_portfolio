```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
schools = len(school_data_complete["school_name"].unique())
students = len(school_data_complete["Student ID"].unique())
budget = school_data["budget"].sum()
math = school_data_complete["math_score"].sum()/students
reading = school_data_complete["reading_score"].sum()/students
#overall_rate = (math+reading)/2
#math_pass_df = school_data_complete.loc[school_data_complete["math_score"] > 70, ["math_score"]]
math_rate = (float((school_data_complete.loc[school_data_complete["math_score"] > 70, ["math_score"]]).count())/students)*100
reading_rate = (float((school_data_complete.loc[school_data_complete["reading_score"] > 70, ["reading_score"]]).count())/students)*100
overall_rate = (math+reading)/2
```

```python
dist_summary = pd.DataFrame({"Total Schools": [schools], 
                             "Total Students": [students], 
                             "Total Budget": [budget], 
                             "Ave Math Score": [math], 
                             "Ave Reading Score": [reading], 
                             "% Math Pass Rate": [math_rate], 
                             "% Reading Pass Rate": [reading_rate],
                             "% Overall Pass Rate": [overall_rate]})


# Format dataframe using map method
dist_summary["Total Students"] = dist_summary["Total Students"].map("{:,}".format)
dist_summary["Total Budget"] = dist_summary["Total Budget"].map("${:,.2f}".format)
dist_summary["Math Pass Rate"] = dist_summary["% Math Pass Rate"].map("{:.2f}".format)
dist_summary["Reading Pass Rate"] = dist_summary["% Reading Pass Rate"].map("{:.2f}".format)

dist_summary
```

```python
grouped_school_data = school_data_complete.groupby(["School ID"])

school_name = school_data_complete["school_name"].unique()

grouped_school_type = grouped_school_data["type"].unique()
#grouped_school_type

grouped_school_students = grouped_school_data["Student ID"].count()
#grouped_school_students

grouped_school_budget = grouped_school_data["budget"].unique()
#grouped_school_budget

grouped_school_budget_per_student = grouped_school_budget/grouped_school_students
#grouped_school_budget_per_student

grouped_school_reading = grouped_school_data["reading_score"].mean()
#grouped_school_reading

grouped_school_math = grouped_school_data["math_score"].mean()
```

```python
district_reading_scores = school_data_complete.loc[school_data_complete["reading_score"]>= 70, :]
district_reading_scores = district_reading_scores.groupby("School ID")
district_reading_scores = district_reading_scores["reading_score"].count()
```

```python
district_math_scores = school_data_complete.loc[school_data_complete["math_score"]>= 70, :]
district_math_scores = district_math_scores.groupby("School ID")
district_math_scores = district_math_scores["math_score"].count()
```

```python
school_math_pass = (district_math_scores/grouped_school_students)*100
school_math_pass
```

```python
school_reading_pass = (district_reading_scores/grouped_school_students)*100
school_reading_pass
```

```python
school_overall = (school_reading_pass + school_math_pass) /2
```

```python
school_summary = pd.DataFrame({"School": school_name,
                               "Type": grouped_school_type,
                               "Students": grouped_school_students,
                               "Total School Budget": grouped_school_budget,
                               "Per Student Budget": grouped_school_budget_per_student,
                               "Ave Math Score": grouped_school_math,
                               "Ave Reading Score": grouped_school_reading,
                               "% Passing Math": school_math_pass,
                               "% Passing Reading": school_math_pass,
                               "% Overall Passing Rate": school_overall})
                               
school_summary.dtypes
school_summary

school_summary["Total School Budget"] = school_summary["Total School Budget"].astype(float)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].astype(float)

school_summary
```
