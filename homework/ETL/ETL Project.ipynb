```python
from sqlalchemy import create_engine
from config import user, password
rds_connection_string =user +":" + password + "@localhost:5432/Nutrition"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
#load data in food_groups table
food_group.to_sql(name='food_groups', con= engine, if_exists='append',index =False)
```

```python
#load data in vitamins table
vitamins.to_sql(name='vitamins', con= engine, if_exists='append',index =False)
```

```python
#load data in minerals table
minerals.to_sql(name='minerals', con= engine, if_exists='append',index =False)
```

```python
#load data in basic_nutirent table
basic_nutrient.to_sql(name='basic_nutrient', con= engine, if_exists='append',index =False)
```

```python
# best calcium in mgs , products: Suggested DV for human body per 100gms of the food
pd.read_sql_query('select fg.id, fg.foodgroup, fg.shortdescrip, m.calcium  from food_groups as fg inner join minerals as m on fg.id =m.id where m.calcium between 900 and 1300 order by m.calcium desc ', con= engine).head(15)
```

```python
# best vitamin C in mgs , products: Suggested DV for human body per 100gms of the food
pd.read_sql_query('select fg.id, fg.foodgroup, fg.shortdescrip, v.vit_c  from food_groups as fg inner join vitamins as v on fg.id =v.id where v.vit_c between 65 and 95 order by v.vit_c desc', con= engine).head(15)
```

```python
# best protien in gms,  products: Suggested DV for human body per 100gms of the food
pd.read_sql_query('select fg.id, fg.foodgroup, fg.shortdescrip, bn.protein  from food_groups as fg inner join basic_nutrient as bn on fg.id =bn.id where bn.protein between 30 and 75 order by bn.protein desc ', con= engine).head(15)
```

```python
# Merge all the tables in one
pd.read_sql_query('select fg.id, fg.foodgroup, fg.shortdescrip,bn.energ_kcal, bn.protein, bn.fat, bn.water, v.vit_c, v.vit_a_iu,v.vit_b6,v.vit_b12,v.vit_c,v.vit_d,v.vit_e, m.calcium,m.phosphorus,m.potassium,m.sodium,m.magnesium from food_groups as fg inner join minerals as m on fg.id =m.id inner join vitamins as v on v.id= fg.id inner join basic_nutrient as bn on bn.id =fg.id', con= engine).head()
```
